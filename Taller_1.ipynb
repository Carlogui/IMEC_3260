## Problema 4.8
A supersonic nozzle is also a convergent–divergent duct, which is fed by a large reservoir at the inlet to the nozzle. In the reservoir of the nozzle, the pressure and temperature are 10 atm and 300 K, respectively. At the nozzle exit, the pressure is 1 atm. Calculate the temperature and density of the flow at the exit. Assume that the flow is isentropic and (of course) compressible.



**Datos:** $p_0 = 10~\text{atm}$, $T_0 = 300~\text{K}$, $p_e = 1~\text{atm}$, $\gamma = 1.4$, $R = 287~\text{J/(kg\cdot K)}$.

Para flujo isentrópico de gas ideal:

$$
\frac{T_e}{T_0}=\left(\frac{p_e}{p_0}\right)^{\frac{\gamma-1}{\gamma}}
\qquad \Rightarrow \qquad
T_e = T_0\left(\frac{p_e}{p_0}\right)^{\frac{\gamma-1}{\gamma}}
$$

Y la densidad a la salida (gas ideal):

$$
\rho_e=\frac{p_e}{R\,T_e}
$$

```python
# Problema 4.8 — Cálculo en SI

# Datos
p0_48_atm = 10.0
T0_48 = 300.0
pe_48_atm = 1.0

# Constantes
gamma_48 = 1.4
R_48 = 287.0          # J/(kg·K)
atm_to_Pa_48 = 101325.0

# Conversiones a Pa
p0_48 = p0_48_atm * atm_to_Pa_48
pe_48 = pe_48_atm * atm_to_Pa_48

# Isentrópico:
Te_48 = T0_48 * (pe_48 / p0_48) ** ((gamma_48 - 1) / gamma_48)

# Gas ideal:
rho_e_48 = pe_48 / (R_48 * Te_48)

# Respuesta
display(Markdown(
    f"**Respuesta:** $T_e = {Te_48:.2f}\\,\\text{{K}}$ y "
    f"$\\rho_e = {rho_e_48:.3f}\\,\\text{{kg/m}}^3$."
))
```

Se busca una expresión para la velocidad de salida $V_e$ en términos de la razón de presiones $p_0/p_e$ y la temperatura del reservorio $T_0$.

**Supuestos:** flujo 1D, estacionario, adiabático, sin trabajo, sin fricción (isentrópico), gas ideal con $\gamma$ constante, y $V_0 \approx 0$ en el reservorio.

---

### 1) Ecuación de energía (reservorio $\to$ salida)

Para un flujo adiabático sin trabajo:
$$
h_0 = h_e + \frac{V_e^2}{2}
$$

Con gas ideal y $h=c_pT$:
$$
c_pT_0 = c_pT_e + \frac{V_e^2}{2}
$$

Despejando $V_e$:
$$
V_e=\sqrt{2c_p\left(T_0-T_e\right)}
$$

---

### 2) Relación isentrópica $T$–$p$

Para flujo isentrópico de gas ideal:
$$
\frac{T_e}{T_0}=\left(\frac{p_e}{p_0}\right)^{\frac{\gamma-1}{\gamma}}
\qquad\Rightarrow\qquad
T_e=T_0\left(\frac{p_e}{p_0}\right)^{\frac{\gamma-1}{\gamma}}
$$

---

### 3) Sustitución y forma final

Sustituyendo $T_e$ en la expresión de $V_e$:
$$
V_e=
\left(
2c_pT_0
\left[
1-\left(\frac{p_e}{p_0}\right)^{\frac{\gamma-1}{\gamma}}
\right]
\right)^{1/2}
$$

Usando $c_p=\dfrac{\gamma R}{\gamma-1}$:
$$
\boxed{
V_e=
\left(
\frac{2\gamma}{\gamma-1}RT_0
\left[
1-\left(\frac{p_e}{p_0}\right)^{\frac{\gamma-1}{\gamma}}
\right]
\right)^{1/2}
}
$$

En términos de $p_0/p_e$:
$$
\boxed{
V_e=
\left(
\frac{2\gamma}{\gamma-1}RT_0
\left[
1-\left(\frac{1}{p_0/p_e}\right)^{\frac{\gamma-1}{\gamma}}
\right]
\right)^{1/2}
}
$$

## Problema 4.12
A supersonic transport is flying at a velocity of 1500 mi/h at a standard altitude of 50,000 ft. The temperature at a point in the flow over the wing is 793.32°R. Calculate the flow velocity at that point.

## Flujo compresible (energía / temperatura de estancamiento)


### 1) Temperatura estándar a 50,000 ft (Apéndice B — unidades americanas)

De la atmósfera estándar (tabla), a $50{,}000~\text{ft}$:
$$
T_\infty \approx 389.97~^\circ\text{R}
$$

---

### 2) Temperatura de estancamiento (flujo adiabático, isentrópico)

Para flujo adiabático sin trabajo y sin pérdidas, la **temperatura de estancamiento** es constante a lo largo de una línea de corriente:
$$
T_0 = T + \frac{V^2}{2c_p}
$$

Aplicando entre el flujo libre y el punto sobre el ala:
$$
T_0 = T_\infty + \frac{V_\infty^2}{2c_p}
\qquad\text{y}\qquad
T_0 = T_2 + \frac{V_2^2}{2c_p}
$$

Igualando y despejando $V_2$:
$$
V_2=\sqrt{2c_p\,(T_0-T_2)}
$$

(Para aire ideal: $c_p=\dfrac{\gamma R}{\gamma-1}$ con $\gamma=1.4$.)

> Nota: Si $T_2$ resulta prácticamente igual a $T_0$, entonces $V_2 \approx 0$ (punto de estancamiento).  

```python
# Problema 4.12 — Cálculo (variables con sufijo _412)
import math

# Datos del enunciado (unidades inglesas dadas)
Vinf_412_mph = 1500.0          # mi/h
T2_412_R = 793.32              # °R

# De la tabla de atmósfera estándar a 50,000 ft (Apéndice B, unidades americanas)
Tinf_412_R = 389.99            # °R

# Constantes aire (SI para evitar líos de unidades)
gamma_412 = 1.4
R_412 = 287.0                  # J/(kg·K)
cp_412 = gamma_412*R_412/(gamma_412 - 1)  # J/(kg·K)

# Conversiones a SI
mph_to_mps = 0.44704
R_to_K = 5/9                   # K = (5/9)*°R

Vinf_412 = Vinf_412_mph * mph_to_mps
Tinf_412 = Tinf_412_R * R_to_K
T2_412 = T2_412_R * R_to_K

# Temperatura de estancamiento a partir del flujo libre
T0_412 = Tinf_412 + Vinf_412**2/(2*cp_412)

# Velocidad en el punto (puede dar negativo por redondeos si T2≈T0)
deltaT_412 = T0_412 - T2_412
V2_412 = math.sqrt(max(0.0, 2*cp_412*deltaT_412))

# Volver a mi/h para reportar también
V2_412_mph = V2_412 / mph_to_mps

display(Markdown(
    f"**Cálculo:** $T_0 = {T0_412*(9/5):.2f}\\,^\\circ\\text{{R}}$.\n\n"
    f"**Respuesta:** $V_2 = {V2_412:.2f}\\,\\text{{m/s}}$ "
    f"$(\\approx {V2_412_mph:.2f}\\,\\text{{mi/h}})$."
))

# (Opcional) Mostrar si es prácticamente estancamiento
if abs(deltaT_412) < 0.5:  # K
    display(Markdown("> Como $T_2 \\approx T_0$, este punto es prácticamente un **punto de estancamiento** ($V_2\\approx 0$)."))
```

## Problema 4.15
A Boeing 747 is cruising at a velocity of 250 m/s at a standard altitude of 13 km. What is its Mach number?


### 1) Definición de número de Mach

$$
M=\frac{V}{a}
$$

donde $a$ es la velocidad del sonido. Para aire ideal:

$$
a=\sqrt{\gamma R T}
$$

---

### 2) Temperatura estándar a 13 km (Apéndice A, SI)

De la tabla de atmósfera estándar (SI) a $h=13~\text{km}$ se obtiene $T$.

---

### 3) Sustitución

$$
M=\frac{V}{\sqrt{\gamma R T}}
$$

```python
# Problema 4.15 — Cálculo (variables con sufijo _415)

# Dato
V_415 = 250.0  # m/s

# De la tabla de atmósfera estándar (Apéndice A, SI) a 13 km:
T_415 = 216.66  # K

# Constantes
gamma_415 = 1.4
R_415 = 287.0  # J/(kg·K)

# Velocidad del sonido y Mach
a_415 = math.sqrt(gamma_415 * R_415 * T_415)
M_415 = V_415 / a_415

display(Markdown(
    f"**Cálculo:** $a = \\sqrt{{\\gamma RT}} = {a_415:.2f}\\,\\text{{m/s}}$.\n\n"
    f"**Respuesta:** $M = {M_415:.3f}$."
))
```

- Pitot: $p_t = 1650$ lb/ft² (presión total o de estancamiento)
- OAT: $T = 500$ °R

### 1) Presión estática a 8000 ft

De la tabla de atmósfera estándar (unidades americanas) a 8000 ft:

$$
p_\infty = 1572.1\ \text{lb/ft}^2
$$

(Usa el valor exacto de tu apéndice.)

---

### 2) Presión dinámica (Pitot, baja velocidad)

$$
q = p_t - p_\infty
$$

---

### 3) Densidad real (con la temperatura dada)

Gas ideal:

$$
\rho = \frac{p_\infty}{R\,T}
$$

con $R = 1716$ ft·lbf/(slug·°R).

---

### 4) Velocidad verdadera (TAS)

Para baja velocidad (incompresible):

$$
q = \frac{1}{2}\rho V^2
\qquad\Rightarrow\qquad
V = \sqrt{\frac{2q}{\rho}}
$$

---

### 5) Velocidad equivalente (EAS)

Definición:

$$
V_E = \sqrt{\frac{2q}{\rho_0}}
$$

donde $\rho_0 = 0.0023769$ slug/ft³ (nivel del mar estándar).

```python
# Datos
h_421_ft = 8000.0
pt_421 = 1650.0          # lb/ft^2 (Pitot = presión total)
T_421 = 500.0            # °R (OAT)

# De la tabla (atmósfera estándar, unidades americanas) a 8000 ft:
p_inf_421 = 1572.1      # lb/ft^2

# Constantes
R_421 = 1716.0           # ft·lbf/(slug·°R)
rho0_421 = 0.0023769     # slug/ft^3 (nivel del mar estándar)
fts_to_mph = 0.6818181818
fts_to_kt  = 0.5924838013

# Dinámica
q_421 = pt_421 - p_inf_421

# Densidad real con OAT
rho_421 = p_inf_421 / (R_421 * T_421)

# TAS (true airspeed)
V_TAS_421 = math.sqrt(2*q_421/rho_421)

# EAS (equivalent airspeed)
V_EAS_421 = math.sqrt(2*q_421/rho0_421)

display(Markdown(
    f"**Resultados:**\n\n"
    f"- Presión dinámica: $q = p_t - p_\\infty = {q_421:.2f}\\,\\text{{lb/ft}}^2$\n"
    f"- Densidad real: $\\rho = {rho_421:.6f}\\,\\text{{slug/ft}}^3$\n\n"
    f"**Respuesta:**\n\n"
    f"- Velocidad verdadera (TAS): $V = {V_TAS_421:.2f}\\,\\text{{ft/s}}$ "
    f"($= {V_TAS_421*fts_to_mph:.1f}\\,\\text{{mi/h}}$, "
    f"$= {V_TAS_421*fts_to_kt:.1f}\\,\\text{{kt}}$)\n\n"
    f"- Velocidad equivalente (EAS): $V_E = {V_EAS_421:.2f}\\,\\text{{ft/s}}$ "
    f"($= {V_EAS_421*fts_to_mph:.1f}\\,\\text{{mi/h}}$, "
    f"$= {V_EAS_421*fts_to_kt:.1f}\\,\\text{{kt}}$)"
))
```

**(ii) Mach detrás del choque:**
$$
M_2^2
=
\frac{1+\frac{\gamma-1}{2}M_1^2}{\gamma M_1^2-\frac{\gamma-1}{2}}
$$

**(iii) Recuperación isentrópica de $M_2$ hasta estancamiento (región subsónica):**
$$
\frac{p_{02}}{p_2}
=
\left(1+\frac{\gamma-1}{2}M_2^2\right)^{\frac{\gamma}{\gamma-1}}
$$

$$
p_{02}
=
p_2\left(1+\frac{\gamma-1}{2}M_2^2\right)^{\frac{\gamma}{\gamma-1}}
$$

---

### 3) Resultado erróneo de Bernoulli (incompresible)

Bernoulli incompresible:

$$
p_0 = p_1+\frac{1}{2}\rho_1 V_1^2
$$

Con gas ideal:

$$
\rho_1 = \frac{p_1}{RT_1}
$$

Y usando $V_1 = M_1 a_1$ y $a_1=\sqrt{\gamma RT_1}$:

$$
V_1 = M_1\sqrt{\gamma RT_1}
$$

```python
# Datos
M1_430 = 2.0
gamma_430 = 1.4
p1_430 = 101325.0  # Pa
T1_430 = 288.15    # K
R_430 = 287.0      # J/(kg·K)

# 1) Presión total isentrópica del flujo
p01_430 = p1_430 * (1 + (gamma_430 - 1)/2 * M1_430**2) ** (gamma_430/(gamma_430 - 1))

# 2) Cuerpo romo: choque normal + desaceleración isentrópica
p2_over_p1_430 = 1 + (2*gamma_430/(gamma_430 + 1)) * (M1_430**2 - 1)
p2_430 = p1_430 * p2_over_p1_430

M2_sq_430 = (1 + (gamma_430 - 1)/2 * M1_430**2) / (gamma_430*M1_430**2 - (gamma_430 - 1)/2)
M2_430 = math.sqrt(M2_sq_430)

p02_430 = p2_430 * (1 + (gamma_430 - 1)/2 * M2_430**2) ** (gamma_430/(gamma_430 - 1))

# 3) Bernoulli incompresible (incorrecto)
rho1_430 = p1_430 / (R_430 * T1_430)
a1_430 = math.sqrt(gamma_430 * R_430 * T1_430)
V1_430 = M1_430 * a1_430
p0_bern_430 = p1_430 + 0.5 * rho1_430 * V1_430**2

# Razones para comparar
ratio_isent_430 = p01_430 / p1_430
ratio_blunt_430 = p02_430 / p1_430
ratio_bern_430 = p0_bern_430 / p1_430

display(Markdown(
    f"**Resultados:**\n\n"
    f"$p_1 = {p1_430}$ Pa\n\n"
    f"$M_1 = {M1_430}$\n\n"
    f"**1) Presión total isentrópica:** $p_{{01}} = {p01_430:.2f}$ Pa, $p_{{01}}/p_1 = {ratio_isent_430:.6f}$\n\n"
    f"**2) Cuerpo romo (choque normal + estancamiento):** $p_{{02}} = {p02_430:.2f}$ Pa, $p_{{02}}/p_1 = {ratio_blunt_430:.6f}$\n\n"
    f"**3) Bernoulli (incorrecto):** $p_0 = {p0_bern_430:.2f}$ Pa, $p_0/p_1 = {ratio_bern_430:.6f}$"
))
```
